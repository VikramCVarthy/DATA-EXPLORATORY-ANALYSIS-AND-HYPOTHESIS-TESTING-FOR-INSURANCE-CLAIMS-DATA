#### 6. Is there any similarity in the amount claimed by males and females?


```python
avg_claim_amt
```

```python
Female_1=mfyisnu.query('gender=="Female"').claim_amount
Male_1=mfyisnu.query('gender=="Male"').claim_amount
```

```python
stats.ttest_ind(Female_1,Male_1)
```

##### Pvalue is > 0.05. Hence we accept Null Hypothesis H0 and There is no relationship for the amount claimed by males and females

### 17. Is there any relationship between age category and segment?

```python
from scipy.stats import chi2_contingency
## Purpose: To test the independence between two categorical variables in a contingency table
# Create a contingency table
contingency_table = pd.crosstab(mfyisnu['Age_grp'], mfyisnu['Segment'])
```

```python
stats.chi2_contingency( contingency_table )
```

#### business conclusion: Pvalue is > 0.05. Hence we accept Null Hypothesis H0 and There is no relationship between age category and segment
